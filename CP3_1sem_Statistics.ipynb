5 – Um dos produtos possui a seguinte função lucro −𝟐𝒑
𝟐 + 𝟓𝟎𝒑 − 𝟏𝟎. Encontre a derivada da
função lucro. Qual seria o melhor preço?

```python
# Definindo a variável p
p = sp.Symbol('p')

p
```

```python
# Definindo a função lucro
funcao_lucro = -2*p**2 + 50*p - 10
funcao_lucro
```

```python
# Encontrando a derivada da função lucro em relação a p
derivada_lucro = sp.diff(funcao_lucro, p)

derivada_lucro
```

```python
# Encontrando o Ponto Máximo (Melhor Preço)

melhor_preco = sp.solve(derivada_lucro)
melhor_preco
```

```python
melhor_preco = sp.solve(derivada_lucro**(1/2))
melhor_preco
print('O melhor preço para o produto é R$', melhor_preco)
```
